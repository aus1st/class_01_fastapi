```python
import requests
from requests.models import Response
from requests.structures import CaseInsensitiveDict

resposne: Response = requests.get('https://jsonplaceholder.typicode.com/todos')

display('Respnse:', resposne)
status_code: int = resposne.status_code
print('Status Code:', status_code)

data: dict = resposne.json()
print('data:', data, type(data))

headers: CaseInsensitiveDict = resposne.headers
print('Response Headers:', headers, type(headers))
```

```python
# Add product
import requests
from requests.models import Response
from requests.structures import CaseInsensitiveDict

product: dict[str,str] = {
    'title': 'BMW Pencil',
    }

resposne: Response = requests.post('https://dummyjson.com/products/add',data=product)

status_code: int = resposne.status_code
print('Status Code:', status_code)

data: dict = resposne.json()
display(data)


headers: CaseInsensitiveDict = resposne.headers
print('Response Headers:', headers, type(headers))
```
